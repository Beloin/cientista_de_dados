# Grafos

```python
from igraph import Graph
from igraph import plot
```

```python
grp = Graph(edges=[(0,1), (1,2), (2,3),(3,0)], directed=True)
grp.vs['label'] = range(grp.vcount())
print(grp)
plot(grp, bbox=(0,0,300,300))
```

```python
grp2 = Graph(edges=[(1,1), (1,2), (2,3), (3,0), (1,1)])
grp2.vs['label'] = range(grp.vcount())
plot(grp2, bbox=(0,0, 300,300))
```

```python
grp2.add_vertex(5)
plot(grp2, bbox=(0,0, 300,300))
```

```python
grp2.add_edge(4,2)
plot(grp2, bbox=(0,0, 300,300))
```
