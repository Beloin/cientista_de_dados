## Faça uma analise expoloratória dos dados do Covid no Brasil Contendo os Seguintes passos:

* Novos Casos
* Incidências

Ánalise os dados do pior para o melhor caso de estados e plote num gráfico para cada problema. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
base = pd.read_csv('HIST_PAINEL_COVIDBR_26out2020.csv', sep=';')
base.shape
```

```python
base.head()
```

```python
todayBase =  base.loc[base['data'] == '2020-10-26']
todayBase.head()
```

```python
todayBase.shape
```

```python
# Aqui Vemos que os Dados são para cidades, precisamos converter em ESTADOS.
todayBase.loc[todayBase['estado'] == 'CE']
```

```python
# Aqui Pegamos todos os dados dos Estados (Pois os estados tem seus munincipios nulos)
estadosBase = todayBase.loc[todayBase['municipio'].isnull()].iloc[1:]
estadosBase.head(5)
```

```python
# De acordo com os dados, as duplicadas reais são as segundas. Então são recuperados os segundos dados
# cleanEstadosBase = estadosBase.loc[estadosBase.duplicated(subset='estado', keep='last')]
cleanEstadosBase = estadosBase.drop_duplicates(subset='estado', keep='first')
cleanEstadosBase.head(5)
```

Então, tendo a limpeza e mineração dos dados anteriormente, agora podemos começar a trabalhar com os dados e implementar o que a questão quer.

```python
# Salvo o DataFrame em uma lista
nomeEstados = cleanEstadosBase.estado.values.tolist()
print('Quantidade de Estados: ', len(nomeEstados))
```

```python
valoresAcumulados = cleanEstadosBase.casosAcumulado.values.tolist()
```

```python
# Exemplo para mostrar todos os dados do Covid.
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111)
plt.bar(nomeEstados,valoresAcumulados, color='grey')
```

```python
# Pegando pelos novos casos:
sortedCasosNovos = cleanEstadosBase.sort_values('casosNovos', ascending=False)
valoresCasosNovos = sortedCasosNovos.casosNovos.values.tolist()
nomeEstados = sortedCasosNovos.estado.values.tolist()

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)
plt.bar(nomeEstados, valoresCasosNovos, color='grey')
```

```python
# Pegando pelos casos acumulados:
sortedCasosAcumulado = cleanEstadosBase.sort_values('casosAcumulado', ascending=False)
valoresCasosAcumulado = sortedCasosAcumulado.casosAcumulado.values.tolist()
nomeEstados = sortedCasosAcumulado.estado.values.tolist()

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)
plt.bar(nomeEstados, valoresCasosAcumulado, color='grey')
```

```python
# Pegando pelos casos de óbito:
sortedObitosAcumulado = cleanEstadosBase.sort_values('obitosAcumulado', ascending=False)
valoresObitosAcumulado = sortedObitosAcumulado.obitosAcumulado.values.tolist()
nomeEstados = sortedObitosAcumulado.estado.values.tolist()

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111)
plt.bar(nomeEstados, valoresObitosAcumulado, color='grey')
```
