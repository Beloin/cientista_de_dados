# Agrupamento usando K-medoids

Os centros dos cluster inicializados não são aleatórios. São pontos reais de dados.

```python
# Importação das bibliotecas
from sklearn import datasets
from sklearn.metrics import confusion_matrix
import numpy as np
from pyclustering.cluster.kmedoids import kmedoids
from pyclustering.cluster import cluster_visualizer
import matplotlib as plt
#pip install pyclustering (executar no Anaconda Prompt)
```

```python
iris = datasets.load_iris()
```

```python
# Visualização dos gráficos do Cluster.
cluster = kmedoids(iris.data[:, 0:2], [14, 20, 30])
## Pontos escolhidos.
cluster.get_medoids()
```

```python
# Aplicação de algoritmos para agrupamento, obtenção das previsões e visualização dos medoides.
cluster.process()
prev = cluster.get_clusters()
medoids = cluster.get_medoids()

# Lista contendo os três agrupamentos que indicam os índices dos dados.
prev
```

```python
v = cluster_visualizer()
v.append_clusters(prev, iris.data[:,0:2])
v.append_cluster(medoids, iris.data[:,0:2], marker='*', markersize=20)
v.show()
plt.show()
```

```python
lista_prev = []
lista_real = []
for i in range(len(prev)):
    for ii in range(len(prev[i])):
        lista_prev.append(i)
        lista_real.append(iris.target[prev[i][ii]])
```

```python
lista_prev = np.asarray(lista_prev)
lista_real = np.asarray(lista_real)
results = confusion_matrix(lista_real, lista_prev)
results
```
