# C-Means
Diferença de KMeans, ele não atribui uma instância de forma absoluta. Utiliza de forma percentual, probalilística.

```python
# Importação das bibliotecas
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import skfuzzy
#pip install -U scikit-fuzzy (executar no Anaconda Prompt)
```

```python
iris = datasets.load_iris()
```

```python
r = skfuzzy.cmeans(data=iris.data.T, c=3, m=2, error=0.005, maxiter=1000, init=None)
```

```python
previsoes_perc = r[1]
```

```python
for x in range(150):
    print(previsoes_perc[0, x], previsoes_perc[1, x], previsoes_perc[2, x])
```

```python
previsoes = previsoes_perc.argmax(axis=0)
resultados = confusion_matrix(iris.target, previsoes)
resultados
```
