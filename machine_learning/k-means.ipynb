# Agrupamento com K-means

Método de agrupamento de dados de acordo com a distância euclidiana a partir de um número de clusters definido pelo usuário.

```python
# Importação das bibliotecas
from sklearn import datasets
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = datasets.load_iris()
# Verificação de quantos registros existem por classe:
unicos, quantidade = np.unique(iris.target, return_counts= True)
print(unicos)
print(quantidade)
```

```python
# Já sabemos que são três Clusters a partir da base de dados 
cluster = KMeans(n_clusters=3)
cluster.fit(iris.data)
```

```python
# Visualização dos trÊs centroids
centroids = cluster.cluster_centers_
centroids
```

```python
# "Previsões" que na verdade são os agrupamentos que retiramos desse dataset
previsoes = cluster.labels_
previsoes
```

```python
unicos2, quantidade2 = np.unique(previsoes, return_counts=True)
print(unicos2)
print(quantidade2)
```

Podemos ver que o agrupamento pode não coincidir com as classes reais.

```python
confusion = confusion_matrix(iris.target, previsoes)
confusion
```

```python
# Geração de Gráfico.
plt.scatter(iris.data[previsoes == 0, 0], iris.data[previsoes == 0, 1], c='green', label='Setosa')
plt.scatter(iris.data[previsoes == 1, 0], iris.data[previsoes == 1, 1], c='red', label='Versicolor')
plt.scatter(iris.data[previsoes == 2, 0], iris.data[previsoes == 2, 1], c='blue', label='Virgica')

plt.legend()
```
