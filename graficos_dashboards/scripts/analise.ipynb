```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn
```

```python
base = pd.read_csv('dados.csv', sep=';')
base.head()
```

```python
base.shape
```

```python
base.isnull().sum()
```

```python
srn.boxplot(base.PIB).set_title('Munincipios por PIB')
```

```python
plt.boxplot(base.PIB, showfliers=False, vert=False, patch_artist=True)
plt.title('Munincipios')
plt.xlabel('PIB')
```

```python
srn.distplot(base.PIB, ).set_title('PIB')
```

```python
# Figura multipla -> PIB e VALOREMPENHO
srn.distplot(base.PIB)
```

```python
agrupado = base.sort_values('PIB').head(10)
agrupado = agrupado.iloc[:,[1,3]]
agrupado 
agrupado.plot.bar(x='MUNICIPIO', y='VALOREMPENHO', color='grey')
```

```python
agrupado = base.sort_values('VALOREMPENHO').head(40)
agrupado = agrupado.iloc[:,[1,3]]
agrupado 
agrupado.plot.bar(x='MUNICIPIO', y = 'VALOREMPENHO', color='gray') 
```
