# Testar e Criar a distribuição Normal

```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

## Dados Normais

```python
dados = norm.rvs(size = 1000)
# dados
```

Verificação por meio de um Histograma

```python
plt.hist(dados, bins=20)
plt.title('Dados')
```

Verficação por meio do plt de gráfico

```python
fig, ax = plt.subplots()
stats.probplot(dados, fit=True, plot=ax)
plt.show()
```

Teste de Shapiro

```python
stats.shapiro(dados)
```

## Dados não normais

```python
dados2 = skewnorm.rvs(4, size=1000)
```

```python
plt.hist(
    dados2,
    bins = 20
)
plt.title('Dados')
```

```python
fig, ax = plt.subplots()
stats.probplot(dados2, fit=True, plot=ax)
plt.show()
```

```python
stats.shapiro(dados2)
```
